```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("resluts_forq2.csv")
df.head()
```

```python
df1 = df[['nationality_name','name','overall']]
nation = []
for i in df['nationality_name']:
    if i not in nation:
        nation.append(i)
print(nation)
```

```python
def rankteam():
    list_n = []
    list_s = []
    for i in nation:
        df2 = df1[df1['nationality_name']==i]
        a = df2['overall'].sum()/len(df2)
        list_n.append(i)
        list_s.append("%.2f" %a)
        d = {'Nation':list_n,'Score':list_s}
        d = pd.DataFrame(d)
        res = d.sort_values(by=['Score'],ascending=False)
    res1 = res.reset_index(drop=True)
    res1['Rank'] = range(1,len(nation)+1)
    print(res1)
```

```python
rankteam()
```
