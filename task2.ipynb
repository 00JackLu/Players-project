```python
# convert csv into dataframe
# import pandas module 
import pandas as pd 
import numpy as np
from datetime import datetime
import difflib
    
# making dataframe 
spider = pd.read_csv("spider.csv") 
players = pd.read_csv("players_22.csv") 
# create a new dataframe of result
# results = pd.DataFrame(index = range(len(spider.axes[0])), columns = range(len(spider.axes[1]+len(players.axes[1]))))
results = pd.concat([spider,players],axis=1)
results = results.drop(index=range(results.shape[0]))
```

```python
nation_list = []
# get the all names of 32 world cup teams
for i in range(len(spider.axes[0])):
    if spider.iloc[i][1] not in nation_list:
        nation_list.append(spider.iloc[i][1])
# len(nation_list) 32 teams
players_new = players
players_new = players_new.drop(index=range(players_new.shape[0]))
# append the players_new
nation_list.append("Korea Republic")
for i in range(len(players.axes[0])):
    if players.loc[i,'nationality_name'] in nation_list:
        players_new = players_new.append(players.loc[i], ignore_index=True)
# nation_list1 = []
# # get the all names of 32 world cup teams
# for i in range(len(players_new.axes[0])):
#     if players_new.loc[i,'nationality_name'] not in nation_list1:
#         nation_list1.append(players_new.loc[i,'nationality_name'])
# print(len(nation_list1))
# for i in range(len(nation_list)):
#     if nation_list[i] not in nation_list1:
#         print(nation_list[i])
# Korea Republic South Korea
# # katar doesn't exist in the file of players_22, so we ignore it(by Ryan)
```

```python
spider['nationality_name'].replace("South Korea", "Korea Republic", inplace=True)# have to direct a column or row
#we must unify the DOB as the same
for i in range(len(spider.axes[0])):
    date_string = spider.iloc[i][3]
    # Parse the date string using the strptime method
    date = datetime.strptime(date_string, '%d.%m.%Y')
    # Use the strftime method to convert the date to a unified format
    unified_date_string = date.strftime('%Y-%m-%d')
    spider.at[i,'birth'] = str(unified_date_string)
    # print(spider.iloc[i][3])
```

```python
count = 0
unknown = spider
unknown = unknown.drop(index=range(unknown.shape[0]))
# create a list to make the useage of fuzzy query
long_name_list = players_new['long_name'].tolist()
short_name_list = players_new['short_name'].tolist()
for i in range(len(spider.axes[0])):
    print(i)
    for ii in range(len(players_new.axes[0])):
        if (spider.iloc[i][1] == players_new.loc[ii,'nationality_name']):#all the names of columns are lower cases
            if (spider.iloc[i,2] == players_new.loc[ii,'long_name']):
                results.loc[i] = pd.concat([spider.iloc[i],players_new.iloc[ii]])
                print("found one")
                count+=1
                break
            elif (spider.iloc[i][3] == players_new.loc[ii,'dob']) and ((spider.iloc[i][4] == players_new.loc[ii,'height_cm']) and (spider.iloc[i][5] == players_new.loc[ii,'weight_kg'])):
                results.loc[i] = pd.concat([spider.iloc[i],players_new.iloc[ii]])
                print("found one")
                count+=1
                break
            else:
                possible_matches = long_name_list
                matches = difflib.get_close_matches(spider.iloc[i,2], possible_matches)
                if(matches):
                    # Find the position of name in the list
                    position = long_name_list.index(matches[0])
                    results.loc[i] = pd.concat([spider.iloc[i],players_new.iloc[position]])
                    print("found one")
                    count+=1
                    break
                else:
                    possible_matches = short_name_list
                    matches = difflib.get_close_matches(spider.iloc[i,2], possible_matches)
                    # Find the position of name in the list
                    if (matches):
                        position = short_name_list.index(matches[0])
                        results.loc[i] = pd.concat([spider.iloc[i],players_new.iloc[position]])
                        print("found one")
                        count+=1
                        break
                    else:
                        print("unknown")
                        unknown.loc[len(unknown.index)] = spider.iloc[i]
                        break
print(count)
```
